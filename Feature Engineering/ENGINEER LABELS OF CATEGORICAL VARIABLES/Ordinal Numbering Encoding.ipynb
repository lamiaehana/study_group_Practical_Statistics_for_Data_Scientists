categorical vars which categories can be meaningfully ordred are called ordinal:

for example: 
Stuent's grade in exam (A, B, C or Fail)
Days of the week can be ordinal with Monday=1 and Sunday=7

when the categorical variable is ordinal, the most straightforward approach is to replace the labels by some ordinal number

```python
import pandas as pd
import datetime
```

```python
base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(0, 30) ]
df = pd.DataFrame(date_list)
df.columns = ['day']
df
```

```python
df['day_of_week'] = df['day'].dt.day_name()
df.head()
```

```python
# enginner categorical variable by ordinal number replacement 

weekday_map = {
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7,
}
```

```python
df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head()
```
