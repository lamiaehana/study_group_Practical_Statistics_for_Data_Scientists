**Target Guided Ordinal Encoding:**  in this technique we will take help of our target variable to encode the categorical data

```python
import pandas as pd

from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
data = pd.read_csv("/content/drive/MyDrive/Feature Engineering/titanic_train.csv", usecols=['Cabin', 'Survived'])
data.head()
```

```python
# fill Na values with an additional label 

data.Cabin.fillna('Missing', inplace=True)
data.head()
```

```python
len(data.Cabin.unique())
```

```python
data['Cabin'] =  data['Cabin'].astype(str).str[0]
data.head()
```

```python
data.Cabin.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['Cabin', 'Survived']], data.Survived, test_size=0.3, random_state=0)

X_train.shape, X_test.shape
```

```python
X_train.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
X_train.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordered_labels = X_train.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_label = {k:i for i, k in enumerate(ordered_labels, 0)}
ordinal_label
```

```python
## replace the labels with the ordred numbers 
X_train['Cabin_ordered'] = X_train.Cabin.map(ordinal_label)
X_test['Cabin_ordered'] = X_test.Cabin.map(ordinal_label)
```
