# One Hot Encoding

consist of replacing the categorical variable by different boolean variables, which take value 0 or 1 to indicate whether or not a certain category / label of the variable was present for that observation

```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/Feature Engineering/titanic_train.csv", usecols=['Sex'])
data.head()
```

# Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
CATEGORICAL_FEATURES = ['Sex'] # CATEGORICAL_FEATURES = ['Pclass', 'Sex', 'Embarked','CabinLetter']
cat_df = data[CATEGORICAL_FEATURES].copy()
cat_df
```

```python
one_hot_encoder = OneHotEncoder(drop='first')
cat_features = one_hot_encoder.fit_transform(cat_df).todense()

pd.DataFrame(cat_features, columns=one_hot_encoder.get_feature_names())
```
