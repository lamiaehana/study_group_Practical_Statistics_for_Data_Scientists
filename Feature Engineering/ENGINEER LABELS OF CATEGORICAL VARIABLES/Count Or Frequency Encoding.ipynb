replacing features with cardinality by frequency or count

Another way to refer to variables that have a multitude of categories, is to call them variables with hight cardinality

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

data = pd.read_csv("/content/drive/MyDrive/Feature Engineering/mercedesbenz.csv", usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'y'])
data.head()
```

```python
for col in data.columns[1:]:
    print(col, ':', len(data[col].unique()), ' labels')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['X1', 'X2', 'X3', 'X4', 'X5', 'X6']], data.y, test_size=0.3, random_state=0)

X_train.shape, X_test.shape
```

```python
X_train.X2.value_counts().to_dict()
```

```python
# create dictionary that maps each label to the counts
X_frequent_map = X_train.X2.value_counts().to_dict()

# replace X2 labels both in the train and test set with the same map
X_train.X2 = X_train.X2.map(X_frequent_map)
X_test.X2 = X_test.X2.map(X_frequent_map)
```
